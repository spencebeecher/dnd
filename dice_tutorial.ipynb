```python
from dnd import Dice
```

```python
Dice(1,6)
```

```python
d = Dice(1,6)

print('Do a single roll of the dice:', d.roll())
print('Simulate 10 rolls:', d.roll(10))
print('Expected value of the dice:', d.expectation())
print('Simulate 10000 rolls, averageing results:', d.roll(10000).mean())
```

```python
d = Dice(1,6) + 4 - Dice(1,8) + Dice(1,6) - 2 + 2 * Dice(1,4)
print(d) 
print(d.expectation(), d.roll(10000).mean())

import pandas as pd
import matplotlib.pyplot as plt
pd.Series(d.roll(10000)).hist(bins=range(1, 22), align='left', rwidth=0.9)
plt.title('Histogram of 10,000 rolls of [1d6 + 4 - 1d8 + 1d6 - 2 + 2d4]')
```
